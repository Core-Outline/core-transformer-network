```python
from transformers import pipeline
import pandas as pd
```

```python
qa_model = pipeline("question-answering")
```

```python
question = "How is growth cost determined"
context = "Growth cost is an important metric to measure the growth and profitability of your business. Businesses measure growth in multiple ways: monthly, yearly and quarterly."
qa_model(question = question, context = context)
```

```python
def is_heading(paragraph):
    # Check if the paragraph is formatted as a heading
    return paragraph.style.name.startswith('Heading')
```

```python
file_path = '../Data/50_Metrics_All_Startup_Founders_Need_to_Know_1684228625.docx'

# Create a Document object to open the DOCX file
doc = docx.Document(file_path)

headings = []
paragraphs = []
para_arr = [x for x in doc.paragraphs]
p = ''
h = ''
for para in doc.paragraphs:
    is_heading = para.style.name.startswith('Heading')
    print(para.text)
    print(is_heading)
    if is_heading:
        if(p != '') : paragraphs.append(p)
        p = ''
        headings.append(para.text)
    else:
        p = p + para.text
        continue
        
        
# Print the extracted headings and paragraphs
print("Headings:")
for heading in headings:
    print(heading)

print("\nParagraphs:")
for paragraph in paragraphs:
    print(paragraph)
```
